```python
import cv2
import matplotlib.pyplot as plt
import numpy
import math

#code diatas berfungsi untuk mengimport library-lobary yang akan digunakan dalam program
#dan ada yang diberikan nama alias dengan keyword as untuk penyingkatan penulisan
```

```python
#membaca dan memanggil gambar.jpg yang disimpan pada variabel image_plt dengan menggunakan library matplotlib dan function imread
image_plt = plt.imread('gambar.jpg')
#membaca dan memanggil gambar.jpg yang disimpan pada variabel image_cv2 dengan menggunakan library cv2 dan function imread
image_cv2 = cv2.imread('gambar.jpg')
#konversi warna gambar image_cv2 dari RGB ke grayscale image(hitam putih) dengan menggunakan library cv2
imagebaru = cv2.cvtColor(image_cv2, cv2.COLOR_BGR2GRAY)
```

1 TRANSLASI (x = -40, y = 30)

```python
#mendapatkan nilai tinggi dan lebar (dimensi) dari imagebaru dengan fungsi shape
(h,w) = imagebaru.shape[:2]
#mendapatkan matrix array dari imagebaru yang akan disimpan pada matrix_image_cv2
matrix_image_cv2 = numpy.array(imagebaru)
#membuat matriks yang berisi nol dengan ukuran 256x256
translasi_baru = numpy.zeros((256,256))
#membuat perulangan dari 0 sampai panjang imagebaru-40
for i in range (len(imagebaru)-40):
    #membuat perulangan dari 0 sampai panjang imagebaru-30
    for j in range (len(imagebaru)-30):
        #mengurangi nilai i dengan -40 dan dimasukan ke dalam variabel x 
        #sehingga akan bergeser searah sumbu x ke kiri
        x = i - (-40)
        #mengurangi nilai i dengan 30 dan dimasukan ke dalam variabel y
        #sehingga akan bergeser searah sumbu y ke bawah
        y = j - 30
        #membuat kondisi jika 0 kurang dari dimensi tinggi dari gambar dan y berada lebih besar dari 0 dan kurang dari nilai dimensi lebar dari gambar
        if 0 < h and 0 < y < w:
            #maka matriks dari imagebaru akan disimpan ke translasi_baru dengan indeks x,y
            translasi_baru[x,y] = imagebaru[i][j]
            
#memanggil library matplotlub function subplot untuk membuat tabel dengan baris berjumlah 1 dan kolom berjumlah 2 dan urutan ke-1
plt.subplot(1,2,1)
#menampilkan gambar image_plt
plt.imshow(imagebaru, cmap="gray")
#menampilkan keterangan gambar yang terdiri dari label, ukuran font dan warna
plt.title(label="ORIGINAL IMAGE", fontsize=10, color="green")
#memanggil library matplotlub function subplot untuk membuat tabel dengan baris berjumlah 1 dan kolom berjumlah 2 dan urutan ke-2
plt.subplot(1,2,2)
#menampilkan gambar imagebaru dengan hanya menggunakan warna hitam dan putih
plt.imshow(translasi_baru, cmap="gray")
#menampilkan keterangan gambar yang terdiri dari label, ukuran font dan warna
plt.title(label="IMAGE TRANSLATION", fontsize=10, color="green")
#menampilkan dimensi dari gambar(tinggi dan lebar)
```

2 ROTATE (170 deg)

```python
#mendapatkan nilai tinggi dan lebar (dimensi) dari imagebaru dengan fungsi shape
(h,w) = imagebaru.shape[:2]
#membuat function rotasi dengan parameter besarnya nilai yang akan dirotasi(derajat)
def rotasi(degree):
    #membuat matriks nol dengan ukuran mengikuti dimensi dari imagebaru (h,w) yang disimpan dalam variabel rotasi
    rotasi = numpy.zeros((w,h))
    #mengunakan rumus minus derajat dikali nilai phi (3,14...) dibagi 180 
    angle = -degree*(numpy.pi/180)
    #mendapatkan nilai tengah dari lebar gambar yang akan dibulatkan ke bawah sebagai center x yang akan menjadi acuan dalam merotasi
    center_x = w // 2
    #mendapatkan nilai tengah dari tinggi gambar yang akan dibulatkan ke bawah sebagai center y yang akan menjadi acuan dalam merotasi
    center_y = h // 2
    #melakukan perulangan sebanyak nilai w (dimensi lebar)
    for i in range(w):
    #melakukan perulangan sebanyak nilai h (dimensi tinggi)
        for j in range(h):
            #menggunakan rumus rotasi dimana nilai indeks i dikurangi center x kemudian dikalikan dengan sin dan cos dari derajat yang dimasukan ke dalam fungsi kemudian dilakukan penambahan dan penguarangan dengan nilai center x dan center y
            x = int((i - center_x) * numpy.cos(angle) - (j - center_y) * numpy.sin(angle) + center_x)
            y = int((i - center_x) * numpy.sin(angle) + (j - center_y) * numpy.cos(angle) + center_x)
            #membuat pengkondisian jika nilai x dan y kurang dari nilai w dan h serta sama atau lebih dari 0 makan 
            if 0 <= x < w and 0 <= y< h:
                #nilai dari indeks x dan y image baru disimpan dalam variabel rotasi mengikuti indeks i dan j
                rotasi[i][j] = imagebaru[x][y]
                  
    #memanggil library matplotlub function subplot untuk membuat tabel dengan baris berjumlah 1 dan kolom berjumlah 2 dan urutan ke-1
    plt.subplot(1,2,1)
    #menampilkan gambar image_plt
    plt.imshow(imagebaru, cmap="gray")
    #menampilkan keterangan gambar yang terdiri dari label, ukuran font dan warna
    plt.title(label="ORIGINAL IMAGE", fontsize=10, color="green")
    #memanggil library matplotlub function subplot untuk membuat tabel dengan baris berjumlah 1 dan kolom berjumlah 2 dan urutan ke-2
    plt.subplot(1,2,2)
    #menampilkan gambar imagebaru dengan hanya menggunakan warna hitam dan putih
    plt.imshow(rotasi, cmap="gray")
    #menampilkan keterangan gambar yang terdiri dari label, ukuran font dan warna
    plt.title(label="ROTATED IMAGE", fontsize=10, color="green")
    #menampilkan dimensi dari gambar(tinggi dan lebar)
    
#memanggil function rotasi dengan memasukan parameter 170
rotasi(170)
```

4 DILATASI (skala = 2)

```python
#membuat function dilated dengan parameter citra dan nilai scalar
def dilated (citra, scalar):
    #mendapatkan dimensi tinggi dan lebar dari citra
    (h,w) = citra.shape[:2]
    #menghitung nilai tinggi dikali scalar yang hasilnya akan dibulatkan ke bawah menggunakan math.floor
    new_H = math.floor(h*scalar)
    #menghitung nilai lebar dikali scalar yang hasilnya akan dibulatkan ke bawah menggunakan math.floor
    new_W = math.floor(w*scalar)
    
    #membuat array nol dengan dimensi mengikuti nilai new_H dan new_Y
    result = numpy.zeros([new_H, new_W])
    
    #membuat perulangan baris sampai nilai new_H
    for i in range (new_H):
    #membuat perulangan kolom sampai nilai new_Y
        for j in range (new_W):
            #mengalikan nilai tinggi dari citra dengan baris i kemudian dibagi dengan new_H yang hasilnya akan dibulatkan
            x = int(h*i/new_H)
            #mengalikan nilai lebar dari citra dengan kolom j kemudian dibagi dengan new_W yang hasilnya akan dibulatkan
            y = int(w*j/new_W)
            #menyimpan hasil citra indeks x dan y ke variabel result indeks i dan j
            result[i][j] = citra[x][y]
            
    #mengembalikan nilai result keluar dari function
    return result
```

```python
#memanggil function dilated dengan memasukan nilai imagebaru dan nilai scalar sama dengan 2
dilated_image = dilated(imagebaru, 2)

#memanggil library matplotlub function subplot untuk membuat tabel dengan baris berjumlah 1 dan kolom berjumlah 2 dan urutan ke-1
plt.subplot(1,2,1)
#menampilkan dimensi gambar asli sebelum dilatasi
print("Dimensi Gambar Asli : ",imagebaru.shape)
#menampilkan gambar image_plt
plt.imshow(imagebaru, cmap="gray")
#menampilkan keterangan gambar yang terdiri dari label, ukuran font dan warna
plt.title(label="ORIGINAL IMAGE", fontsize=10, color="green")
#memanggil library matplotlub function subplot untuk membuat tabel dengan baris berjumlah 1 dan kolom berjumlah 2 dan urutan ke-2
plt.subplot(1,2,2)
#menampilkan dimensi gambar setalah dilatasi
print("Dimensi Gambar Dilatasi : ",dilated_image.shape)
#menampilkan gambar imagebaru dengan hanya menggunakan warna hitam dan putih
plt.imshow(dilated_image, cmap="gray")
#menampilkan keterangan gambar yang terdiri dari label, ukuran font dan warna
plt.title(label="IMAGE DILATATION", fontsize=10, color="green")
#menampilkan dimensi dari gambar(tinggi dan lebar)
```

PERBANDINGAN HISTOGRAM GRAYSCALE AND NEGATIVE IMAGE

```python
#membuat function cari_ni dengan parameter citra
def cari_ni(citra):
    #membuat array nol dengan panjang 256
    histogram_value = numpy.zeros(256)
    #mendapatkan dimensi tinggi dan lebar dari parameter citra yang dimasukan 
    w,h = citra.shape
    #membuat perulangan baris sampai dengan nilai tinggi
    for i in range(w):
        #membuat perulangan baris sampai dengan nilai lebar
        for j in range(h):
            #kemudian canvas ni dengan indeks citra akan selalu ditambahkan dengan 1 setiap perukangan terjadi
            histogram_value[ citra[i,j] ] += 1
    #mengembalikan nilai historgram_value dari function
    return histogram_value
```
