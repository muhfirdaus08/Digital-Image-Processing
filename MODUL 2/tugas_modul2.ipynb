```python
import cv2
import imutils
import numpy
import matplotlib.pyplot as plt
```

```python
#tampilkan hasil 2 gambar

def show_img2(w, h, citra1, citra2, caption1, caption2):  
    plt.figure(figsize=(w,h))
    plt.subplot(1,2,1)
    plt.title(label=caption1, fontsize=10, color="green")
    plt.imshow(citra1, cmap="gray")
    plt.subplot(1,2,2)
    plt.title(label=caption2, fontsize=10, color="green")
    plt.imshow(citra2, cmap="gray")
```

```python
#tampilkan hasil 3 gambar 

def show_img3(w, h, citra1, citra2, citra3, caption1, caption2, caption3):  
    plt.figure(figsize=(w,h))
    plt.subplot(1,3,1)
    plt.title(label=caption1, fontsize=10, color="green")
    plt.imshow(citra1, cmap="gray")
    plt.subplot(1,3,2)
    plt.title(label=caption2, fontsize=10, color="green")
    plt.imshow(citra2, cmap="gray")
    plt.imshow(citra1, cmap="gray")
    plt.subplot(1,3,3)
    plt.title(label=caption3, fontsize=10, color="green")
    plt.imshow(citra3, cmap="gray")
```

```python
citra_asli = plt.imread('gambar.jpg')
citra = cv2.imread('gambar.jpg')

citra_gray = cv2.cvtColor(citra, cv2.COLOR_BGR2GRAY)
```

```python
citra = cv2.imread('gambar.jpg')
citra_rgb = cv2.cvtColor(citra, cv2.COLOR_BGR2RGB)
#konversi ke gray image
citra_gray = cv2.cvtColor(citra, cv2.COLOR_BGR2GRAY)

show_img2(7, 3, citra_rgb, citra_gray, "ORIGINAL IMAGE", "ROTATED IMAGE")
```

```python
# 1. TRANSLASI

translasi = imutils.translate(citra_gray, -40, 30)

show_img2(7, 3, citra_gray, translasi, "ORIGINAL IMAGE", " TRANSLATION IMAGE")
```

```python
# 2. ROTASI

angle = 170
rotated = imutils.rotate(citra_gray,angle)

show_img2(7, 3, citra_gray, rotated, "ORIGINAL IMAGE", "ROTATED IMAGE")
```

```python
# 3. FLIP 

flip_x = cv2.flip(citra_gray, 0)
flip_y = cv2.flip(citra_gray, 1)

show_img3(10, 3, citra_gray, flip_x, flip_y, "ORIGINAL IMAGE", "FLIP SUMBU X", "FLIP SUMBU Y")
```

```python
# 4. DILATASI (SCALA 2)

w = citra_gray.shape[0] * 2
h = citra_gray.shape[1] * 2

resized = cv2.resize(citra_gray, (w,h), interpolation=cv2.INTER_AREA)

print("Dimensi sebelum dilatasi : ", citra_gray.shape)
print("Dimensi setelah dilatasi : ", resized.shape)


show_img2(7, 3, citra_gray, resized, "ORIGINAL IMAGE", "DILATATION IMAGE")
```

```python
# Mencari Nilai Histogram

def histogram(citra):
    histogram_value = numpy.zeros(256)
    h,w = citra.shape
    for i in range(h):
        for j in range(w):
            histogram_value[ citra[i,j] ] += 1
    return histogram_value

canvas = [i for i in range(256)]   
```

```python
# 5. NEGATIVE (GRAY)

gray = cv2.imread('gambar.jpg', 0)
gray_negative = abs(255 - gray)

show_img2(7, 3, gray, gray_negative, "GRAY IMAGE", "GRAY - NEGATIVE IMAGE")
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
hist_citra = histogram(gray)
plt.bar(canvas, hist_citra)
plt.title(label="GRAY IMAGE", fontsize=10, color="green")
plt.subplot(1,2,2)
hist_negative = histogram(gray_negative)
plt.bar(canvas, hist_negative)
plt.title(label="GRAY - NEGATIVE", fontsize=10, color="green")
```

```python
plt.subplot(1,1,1)
plt.bar(canvas, hist_citra)
plt.subplot(1,1,1)
plt.bar(canvas, hist_negative)
plt.title(label="GRAY IMAGE (BLUE) | GRAY - NEGATIVR (ORANGE)", fontsize=10, color="green")
```
