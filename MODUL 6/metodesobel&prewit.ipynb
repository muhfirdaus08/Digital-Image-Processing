```python
import PIL.Image as np
import matplotlib.pyplot as np1
import numpy as np3
```

```python
gambar = np.open("gambar.jpg").convert("L")

pil = np1.imshow(gambar, cmap = "gray")
baru = np3.array(gambar)
baru1 = np3.array(gambar)

M,N = baru.shape
print(baru.shape)
```

Menampilkan gambar asli sebelum di eksekusi sehingga nantinya akan diubah ke dalam deteksi tepi sobel & prewitt

```python
sx = np3.array([[-1, 0, 1],
                [-2, 0, 2],
                [-1, 0, 1]
               ])

sy = np3.array([[1, 2, 1],
                [0, 0, 0],
                [-1, -2, -1]
                ])
print(sy)
```

#METODE SOBEL

Menampilkan matriks yang disesuaikan dengan rumus dari sobel

```python
#SOBEL
def sobel(matriks, kernel):
    arr_tam = np3.zeros(matriks.shape)
    for i in range(1, matriks.shape[0]-1):
        for j in range(1, matriks.shape[1]-1):  
            x_kernel = 0
            for k in range(i-1, i+2):
                y_kernel = 0
                for l in range(j-1, j+2):
                    arr_tam[i][j] += matriks[k][l] * kernel[x_kernel][y_kernel]
                    y_kernel+=1
                x_kernel+=1
    return arr_tam
# arr_tam[i][j] = matriks[i-1][j-1] * kernel[0][0] + matriks[i-1][j] * kernel[0][1] + matriks[i-1][j+1] * kernel[0][2]
#             + matriks[i][j-1] * kernel[1][0] + matriks[i][j] * kernel[1][1] + matriks[i][j+1] * kernel[1][2]
#             + matriks[i+1][j-1] * kernel[2][0] + matriks[i+1][j] * kernel[2][1] + matriks[i+1][j+1] * kernel[2][2]
#     np1.imshow(arr_tam, cmap="gray")
```

Pada kode diatas dimaksudkan untuk menghitung dari sobel itu sendiri dengan melalui langkah-langkah yang tertera pada kode diatas. Adapun langkah-langkah tersebut dilakukan dengan mengguanakan perulangan serta menggunakan kernel dan hasil tersebut nantinya akan dihitung dan nilainya akan dikembalikan

```python
kon1 = sobel(baru,sx)
np1.imshow(sobel(baru, sx), cmap="gray")
```

Menampilkan gambar yang sudah dilakukan deteksi tepi sobel, hal ini merupakan hasil konvolusi pertama

```python
kon2 = sobel(baru,sy)
np1.imshow(sobel(baru, sy), cmap="gray")
```

```python
#Menampilkan gambar yang sudah dilakukan deteksi tepi sobel, hal ini merupakan hasil konvolusi kedua
```

```python
akhir = abs(kon1) + abs(kon2)
np1.imshow(akhir, "gray")
```

Menampilkan gambar yang sudah dilakukan deteksi tepi sobel, hal ini merupakan hasil konvolusi akhir

#METODE PREWITT

```python
#PREWITT
sx = np3.array([[-1, 0, 1],
                [-1, 0, 1],
                [-1, 0, 1]
               ])

sy = np3.array([[-1, -1, -1],
                [0, 0, 0],
                [1, 1, 1]
                ])
print(sy)
```

Menampilkan matriks yang disesuaikan dengan rumus dari prewitt

```python
def prewitt(matriks, kernel):
    arr_tam = np3.zeros(matriks.shape)
    for i in range(1, matriks.shape[0]-1):
        for j in range(1, matriks.shape[1]-1):  
            x_kernel = 0
            for k in range(i-1, i+2):
                y_kernel = 0
                for l in range(j-1, j+2):
                    arr_tam[i][j] += matriks[k][l] * kernel[x_kernel][y_kernel]
                    y_kernel+=1
                x_kernel+=1
    return arr_tam
# arr_tam[i][j] = matriks[i-1][j-1] * kernel[0][0] + matriks[i-1][j] * kernel[0][1] + matriks[i-1][j+1] * kernel[0][2]
#             + matriks[i][j-1] * kernel[1][0] + matriks[i][j] * kernel[1][1] + matriks[i][j+1] * kernel[1][2]
#             + matriks[i+1][j-1] * kernel[2][0] + matriks[i+1][j] * kernel[2][1] + matriks[i+1][j+1] * kernel[2][2]
#     np1.imshow(arr_tam, cmap="gray")
```

Pada kode diatas dimaksudkan untuk menghitung dari prewitt itu sendiri dengan melalui langkah-langkah yang tertera pada kode diatas. Adapun langkah-langkah tersebut dilakukan dengan mengguanakan perulangan serta menggunakan kernel dan hasil tersebut nantinya akan dihitung dan nilainya akan dikembalikan. Langkah prewitt ini sama dengan langkah pada deteksi tepi sobel, yang membedakan hanya pada matriksnya.

```python
kon1 = prewitt(baru,sx)
np1.imshow(prewitt(baru, sx), cmap="gray")
```

Menampilkan gambar yang sudah dilakukan deteksi tepi prewitt, hal ini merupakan hasil konvolusi pertama

```python
kon2 = prewitt(baru,sy)
np1.imshow(prewitt(baru, sy), cmap="gray")
```

Menampilkan gambar yang sudah dilakukan deteksi tepi prewitt, hal ini merupakan hasil konvolusi kedua

```python
akhir = abs(kon1) + abs(kon2)
np1.imshow(akhir, "gray")
```
