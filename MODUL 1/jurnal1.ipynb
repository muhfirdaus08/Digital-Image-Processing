```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

#code diatas berfungsi untuk mengimport library-lobary yang akan digunakan dalam program
#dan diberikan nama alias dengan keyword as untuk penyingkatan penulisan
```

#HISTOGRAM

```python
img = plt.imread('./benzema.jpg')

#code diatas akan memanggil library matplotlib.pyplot atau plt yand didalamnya terdapat
#function imread untuk membaca citra sesuai dengan parameter yang dimasukan
#kemudian akan hasilnya akan disimpan dalam variabel img
```

```python
plt.title(label="RGB Image", fontsize=15, color="green")
plt.imshow(img)

#berfungsi untuk menampilkan citra asli yang tersimpan dalam variabel img
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

#code di atas merupakan color space conversion yang berfungsi untuk menkonversi warna dari citra dalam 
#variabel img dari yang 3 warna yaitu warna biru, hijau dan merah menjadi 
#dua warna saja hitam dan putih atau abu-abu
```

```python
plt.title(label="GrayScale Image", fontsize=15, color="green")
plt.imshow(img_gray, cmap="gray")

#berfungsi untuk menampilkan citra dari variabel img_gray dengan diubah
#color mapnya menjadi gray
```

```python
print(img_gray.shape)

#berfungsi untuk menampilkan dimensi (tinggi dan lebar) dari citra img_gray
```

```python
tinggi, lebar = img_gray.shape

#meninisialiasi variabel tinggi dan lebar diisi dengan nilai dari dimensi citra img_gray

print(tinggi, lebar)

#menampilkan nilai dari variabel tinggi dan lebar
```

```python
n = tinggi * lebar

#berfungsi untuk mengetahui banyaknya piksel dari citra img_gray yang sudah diperoleh dari tinggi
#dan lebar dengan mengalikan keduanya dan disimpan dalam variabel n

print(n)

#menampilkan nilai dari variabel n
```

```python
grayValue = [i for i in range(256)]

#berfungsi untuk memberikan nilai derajat keabuan untuk citra yang 8 bit (0-255)
#dengan melakukan perulangan

print(grayValue)

#menampilkan derajat keabuan yang disimpan dari variabel grayValue
```

```python
print(img_gray.min())
print(img_gray.max())

#menampilkan derajat keabuan terendah dan tertinggi dari citra img_gray
```

```python
print(img_gray)

#menampilkan seluruh nilai matriks derajat keabuan dari citra img_gray
```

```python
ni = np.zeros(256).astype(int)

#menginisialisasi variabel ni dengan array yang berisi nol semua sebanyak 256 
#dan bertipe int

for i in range (tinggi):
    for j in range (lebar):
        val = img_gray[i,j]
        ni[val] += 1

#berfungsi untuk mengetahui distribusi derajat keabuan setiap piksel pada citra img_gray dengan 
#melalukan perulangan bersarang yang dimulai dari tinggi citra untuk indeks i
#dan kemudian perulangan lebar citra untuk indeks j
#selanjutnya diinisialisasi variabel val dengan citra img_gray dengan indeks i dan j
#kemudian akan dihitung banyaknya nilai setiap untuk masing-masing nilai pada matriks 
#citra img_gray
```

```python
print(ni.sum())
```

```python
print(ni)

#menampilkan nilai dari variabel ni
```

```python
hi = ni/n

#berfungsi untuk mengetahui nilai hi yang didapatkan dari distribusi derajat keabuan (ni)
#dan dibagi dengan jumlah seluruh piksel (n) dari citra img_gray

print(hi)

#menampilkan nilai dari variabel hi
```

```python
plt.title(label="Histogram", fontsize=15, color="green")
plt.bar(grayValue, ni)

#menampilkan histogram dari citra img_gray dengan memanggil function bar dari library matplotlib 
#yang diisikan parameter derajat keabuan (grayValue) dan istribusi derajat keabuan (ni)
```

#KONVOLUSI CARA 1

```python
kernel = np.array([
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0],
    [-1,5,-1],
])

#variabel kernel merupakan inisialisasi untuk mengimport nilai kernelnya dalam bentuk array yang #berdimensi 4x3
```

```python
konvolusi_noclip = np.zeros(img_gray.shape)

#variabel konvolusi berisi array nol dengan ukuran mengikuti dimensi
#dari citra img_gray. Variabel ini berfungsi untuk memberikan ruang kosong ketika melalukan konvolusi

for row in range (1278):
    for col in range (1021):
        
#for row dan for col in range dengan parameter 1278 dan 1021 yang merupakan jumlah baris dan kolom yang #akan dilakukan konvolusiyang seharusnya 1280 dan 1023 karena
#untuk mengindari terjadinya out of bonds


        konvolusi_noclip[row][col] = img_gray[row][col]*kernel[0][0] + img_gray[row][col+1]*kernel[0][1] + img_gray[row][col+2]*kernel[0][2] +  img_gray[row+1][col]*kernel[1][0] + img_gray[row+1][col+1]*kernel[1][1] + img_gray[row+1][col+2]*kernel[1][2]+ img_gray[row+2][col]*kernel[2][0]+img_gray[row+2][col+1]*kernel[2][1] + img_gray[row+2][col+2]*kernel[2][2]
        
#konvolusi[row][col] akan menampung hasil dari perkalian baris dan kolom antara
#mariks citra img_gray dengan matriks kernel berdasarkan dengan rumus konvolusi dan
#disesuaikan dengan indeksnya
```

```python
print(img_gray.shape)
```

```python
konvolusi_noclip

#menampilkan array dari hasil konvolusi
```

```python
plt.title(label="Convolution Without Clipping", fontsize=15, color="green")
plt.imshow(konvolusi_noclip, cmap="gray")

#menampilkan citra img_gray hasil konvolusi tanpa clipping
```

```python
konvolusi_clip = np.clip(konvolusi_noclip,a_min=0,a_max=255)


#konvolusi_clip merupakan variabel untuk menampung hasil clipping citra
#clipping dilakukan dengan memanggil fungsi clip dari library numpy
#clipping dengan paramater konvolusi_noclip, a_min=0 dimana nilai yang kurang dari nol akan 
#diset menjadi nol dan parameter a_max=255 dimana nilai yang lebih dari 255 akan 
#diset menjadi 255 sehingga gambar dapat terlihat lebih normal
```

```python
plt.title(label="Convolution With Clipping", fontsize=15, color="green")
plt.imshow(konvolusi_clip, cmap="gray")

#menampilkan citra img_gray hasil konvolusi dengan clipping
```
